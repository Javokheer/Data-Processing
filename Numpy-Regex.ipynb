# A fast Find Funtion for a Sorted List

Finds leftmost value greater than x in a sorted list

```python
def find_next(a, x):
    i = bisect.bisect_right(a, x)
    if i < len(a):
        return a[i]
    return False

print(find_next([10, 15, 20, 25, 30],30))
```

# A simple get_grade function
get_grade uses a list of cutoffs to split grades into 5 ranges, then uses the bisect index to retrun the corresponding grade.

```python
def get_grades(score, cutoffs=[60, 70, 80, 90], grades='FDCBA'):
    i = bisect.bisect_right(cutoffs, score)
    return grades[i]

grades = [get_grades(score) for score in [52, 99, 77, 70, 90, 100]]
print(grades)
```

# Goups

```python
#Until now we ahev been talking about a regex as a signle pattern which is matched. To group patterns you use parenthesis, 
# which is actually pretty natural

re.findall('([\w]*)([\n])', wiki)
```

```python
#To get list of match objects we use finditer()
for item in re.finditer('([\w]*)([\n])', wiki):
    print(item.group())
```

```python
# groups () method retruns a tuple of the group. We can get an individual group using group(number), where group(0) is the whole
# match, and each other number is the portion of the match we are interested in. In this case, we want group(1)
for item in re.finditer('([\w]*)([\n])', wiki):
    print(item.group(1))
```

```python
for item in re.finditer('(?P<title>[\w]*)(?P<edit_link>[\w])', wiki):
    print(item.groupdict()['title'])
```

```python
#Verbose mode for python regex. Verbose mode allows to write more readable. For this mode, we have to explicitly indicate all 
# whitespace characters, eigher by prepending them with a / or by using the /s spacial value. However, this means we can write
#our regex a bit more like code, and can even include comments with #

pattern = """
(?P<title>.*)       # university title
(\ located\ in\ )  # an indicator of the location
(?P<city>\w*)       # city the university is in 
(,\ )               # seperator for the state
(?P<state>\w*)      # the state the city is located in"""

# Now we call finditer() we just pass re.VERBOSE flag as the last parameter, this makes it much easier to understand large regexes

for item in re.finditer(pattern,wiki,re.VERBOSE):
    print(item.groupdict())
```
